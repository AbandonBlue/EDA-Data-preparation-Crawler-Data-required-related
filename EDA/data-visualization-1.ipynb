```python
# 介面: 1. MATLAB, 2. 物件導向式



# 1. MATLAB: 狀態形式
# 建立 figure
plt.figure()

# 建立2個panel中的第一個, 設定目前的軸
plt.subplot(2, 1, 1)    # (rows, columns, panel number)
plt.plot(x, np.sin(x))

# 建立2個panel中的第一個, 設定目前的軸
plt.subplot(2, 1, 2)    # (rows, columns, panel number)
plt.plot(x, np.cos(x))
```

## 調整圖表：線條顏色、樣式

```python
plt.plot(x, np.sin(x), color='g')              # 顏色編碼
plt.plot(x, np.sin(x-1), color='blue')         # 名字
plt.plot(x, np.sin(x-2), color='w')
plt.plot(x, np.sin(x-3), color='#FFDD44')
plt.plot(x, np.sin(x-4), color='chartreuse')
plt.plot(x, np.sin(x-4), color='0.75')         # 灰階
plt.plot(x, np.sin(x-5), color=(1.0, 0.2, 0.3))# RGB
```

```python
# 如果沒有顏色指定，會自動用default的組合
# 可以用linestyle參數去設定線條的格式

plt.plot(x, x+0, linestyle='solid')
plt.plot(x, x+1, linestyle='dashed')
plt.plot(x, x+1, linestyle='dashdot')
plt.plot(x, x+2, linestyle='dotted')
plt.plot(x, x+3, linestyle='dashed')

# 簡單指定
plt.plot(x, x+4, linestyle='-')   # solid
plt.plot(x, x+5, linestyle='--')  # dashed
plt.plot(x, x+6, linestyle='-.')  # dashdot
plt.plot(x, x+7, linestyle=':')   # dotted
```

```python
# 如果喜歡簡潔寫法，可以把color、linestylee透過沒有關鍵字方式去用在plt.plot() function 中
plt.plot(x, x+0, '-g')
plt.plot(x, x+1, '--c')
plt.plot(x, x+2, '-.k')
plt.plot(x, x+3, ':r')
```

### 基本用法
- solid
    - '-'
- dashed
    - '--'
- dashdot
    - '-.'
- dotted
    - ':'

## 調整圖表: Axes範圍
- 透過plt.xlim()、plt.ylim()去控制。

```python
plt.plot(x, np.sin(x))

plt.xlim(-1,  11)
plt.ylim(-1.5, 1.5)
```

```python
# 或者基於某些原因想要將一個axis反向顯示化，只要倒過來就可以

plt.plot(x, np.sin(x))

plt.xlim(10, 0)
plt.ylim(1.2, -1.2)
```

```python
# plt.axis() 也可以一次設定x, y的範圍, 傳遞list即可[xmin, xmax, ymin, ymax]

plt.plot(x, np.sin(x))
plt.axis([-1, 11, -1.5, 1.5])
```

```python
# 可以讓圖片更tight

plt.plot(x, np.sin(x))
plt.axis('tight')
```

```python
# 相同比例

plt.plot(x, np.sin(x))
plt.axis('equal')
```

## 為圖表加上標籤

```python
plt.plot(x, np.sin(x))
plt.title('A Sine Curve')
plt.xlabel('x')
plt.ylabel('sin(x)')
```

```python
# 單一axes顯示多線條時，需要加上一些表示去幫助辨認。

plt.plot(x, np.sin(x), '-g', label='sin(x)')   # 如果沒有plt.legend是不會顯示
plt.plot(x, np.cos(x), '-.r', label='cos(x)')
plt.axis('equal')

plt.legend() # 說明
```

```python
# matplotlib的陷阱: 從狀態轉換到物件導向大部分都可以用相同函式，但一些設定是用不同名稱的function

ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(xlim=(0, 10), ylim=(-2, 2), xlabel='x', ylabel='sin(x)', title='A Simple Plot')
```

## 簡單的scatter plot

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
# 可以使用plt.plot

x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, 'o', color='black')  # 'o'
```

```python
# 常用標記

rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x',  '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker, label=f'marker={marker}')

plt.legend()
plt.xlim(0, 1.8)
```

```python
# 字元編碼可以一併設定線條與顏色

plt.plot(x, y, '-ok')  # line -, circle marker, block
```
